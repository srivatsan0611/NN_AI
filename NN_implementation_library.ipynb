```python
import pandas as pd #For handling the data
from keras.models import Sequential
from keras.layers import Dense #NN Module
import numpy as np #For neural network dtype
```

# PRE PROCESSING AND T T SPLIT

```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv(r"housepricedata.csv")
df.info()
```

```python
df
```

Converting into Array

```python
data = df.values
data
```

X and Y :

X will be the "Input layer" -> As in, the factors that will determine the output

Y will be the "Output Layer" -> As in, the factors determine this output 

As a result, we assign 10 columns to variable "X" and last to Y

```python
X = data[:,0:10]
Y = data[:,10]
```

```python
#Proceeding to Scale the Data

min_max_scaler = preprocessing.MinMaxScaler()
X_scale = min_max_scaler.fit_transform(X)
```

```python
X_train, X_val_and_test, Y_train, Y_val_and_test = train_test_split(X_scale, Y, test_size=0.3)
```

```python
X_val, X_test, Y_val, Y_test = train_test_split(X_val_and_test, Y_val_and_test, test_size=0.5) #50% of the DataSet
```

```python
print(X_train.shape, X_val.shape, X_test.shape, Y_train.shape, Y_val.shape, Y_test.shape)
```

# Building the Neural Network

In words, we want to have these layers:

    Hidden layer 1: 32 neurons, ReLU activation
    Hidden layer 2: 32 neurons, ReLU activation
    Output Layer: 1 neuron, Sigmoid activation

```python
model = Sequential([
    Dense(32, activation='sigmoid', input_shape=(10,)),
    Dense(32, activation='sigmoid'),
    Dense(1, activation='sigmoid'),
])
```

```python
model.compile(optimizer='sgd',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```
